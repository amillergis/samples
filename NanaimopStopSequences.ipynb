```python
import pandas as pd
```

```python
stop_times = pd.read_csv(r'stop_times.txt')
trips = pd.read_csv(r'trips.txt')
stops = pd.read_csv(r'stops.txt')
```

```python
stops_stop_times = pd.merge(stop_times,stops,on='stop_id')
trip_stops = pd.merge(stops_stop_times,trips,on='trip_id')
```

```python
trip_stops
```

```python
sorted_trips = trip_stops[[
        'route_id','trip_id','shape_id','direction_code','pattern_code','stop_id','stop_sequence','timepoint','departure_time',
        'stop_name','stop_lat','stop_lon'
    ]].sort_values(by=['route_id','trip_id','pattern_code','stop_sequence'])

def assign_trip_time(group):
    group.trip_id = group.iloc[0].departure_time.replace(':','')[:4]
    return group

sorted_trips.groupby(by='trip_id').apply(assign_trip_time)
```

```python
sorted_trips.to_excel('NanaimoStopSequencesbyTrip.xls')
```
