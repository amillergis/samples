```python
import re
import pandas as pd
from datetime import date
```

# Comox Geocoding

```python
def Comox_Addresses(row, field):
    address = str(row[field]).strip()
    return re.sub('^.*, |[^\d]\d*\w?(?=( COMOX| COURTENAY| CUMBERLAND| Courtenay))','',address,re.IGNORECASE)

sheet1 = pd.read_excel(r"ComoxHandyDART Driver's Manifest Feb 1-8, 2016.xlsx",
                           sheetname=0)
sheet1['Trip_Date'] = sheet1.apply(lambda x: date(2016,2,1), axis=1)
sheet1['Origin Address'] = sheet1.apply(Comox_Addresses,axis=1,args=('Origin Address',))
sheet1['Dest Address'] = sheet1.apply(Comox_Addresses,axis=1,args=('Dest Address',))

sheet2 = pd.read_excel(r"ComoxHandyDART Driver's Manifest Feb 1-8, 2016.xlsx",
                           sheetname=1)
sheet2['Trip_Date'] = sheet2.apply(lambda x: date(2016,2,2), axis=1)
sheet2['Origin Address'] = sheet2.apply(Comox_Addresses,axis=1,args=('Origin Address',))
sheet2['Dest Address'] = sheet2.apply(Comox_Addresses,axis=1,args=('Dest Address',))

sheet3 = pd.read_excel(r"ComoxHandyDART Driver's Manifest Feb 1-8, 2016.xlsx",
                           sheetname=2)
sheet3['Trip_Date'] = sheet3.apply(lambda x: date(2016,2,3), axis=1)
sheet3['Origin Address'] = sheet3.apply(Comox_Addresses,axis=1,args=('Origin Address',))
sheet3['Dest Address'] = sheet3.apply(Comox_Addresses,axis=1,args=('Dest Address',))

sheet4 = pd.read_excel(r"ComoxHandyDART Driver's Manifest Feb 1-8, 2016.xlsx",
                           sheetname=3)
sheet4['Trip_Date'] = sheet4.apply(lambda x: date(2016,2,4), axis=1)
sheet4['Origin Address'] = sheet4.apply(Comox_Addresses,axis=1,args=('Origin Address',))
sheet4['Dest Address'] = sheet4.apply(Comox_Addresses,axis=1,args=('Dest Address',))

sheet5 = pd.read_excel(r"ComoxHandyDART Driver's Manifest Feb 1-8, 2016.xlsx",
                           sheetname=4)
sheet5['Trip_Date'] = sheet5.apply(lambda x: date(2016,2,5), axis=1)
sheet5['Origin Address'] = sheet5.apply(Comox_Addresses,axis=1,args=('Origin Address',))
sheet5['Dest Address'] = sheet5.apply(Comox_Addresses,axis=1,args=('Dest Address',))

concat_df = pd.concat((sheet1,sheet2,sheet3,sheet4,sheet5))
concat_df.reset_index(drop=True).to_csv('Comox.csv',index_label='Trip_Number')
```

# Kelowna Geocoding

```python
def Kelowna_Addresses(row, field):
    address = str(row[field]).strip()
    address = re.sub('^.*, | \d*$|\d* (WEST|EAST)','',address,re.IGNORECASE)
    if field == 'ADDRESS':
        address += ' {0}'.format(row['CITY'])
    elif field == 'ADDRESS.1':
        address += ' {0}'.format(row['CITY.1'])
    return address

sheet1 = pd.read_excel(r"Kelowna-Matt's Report II.xlsx",
                           sheetname=0,header=1, skiprows=(2,))
sheet1['Trip_Date'] = sheet1.apply(lambda x: date(2016,2,1), axis=1)
sheet1['Origin Address'] = sheet1.apply(Kelowna_Addresses,axis=1,args=('ADDRESS',))
sheet1['Dest Address'] = sheet1.apply(Kelowna_Addresses,axis=1,args=('ADDRESS.1',))

sheet2 = pd.read_excel(r"Kelowna-Matt's Report II.xlsx",
                           sheetname=1,header=1, skiprows=(2,))
sheet2['Trip_Date'] = sheet2.apply(lambda x: date(2016,2,2), axis=1)
sheet2['Origin Address'] = sheet2.apply(Kelowna_Addresses,axis=1,args=('ADDRESS',))
sheet2['Dest Address'] = sheet2.apply(Kelowna_Addresses,axis=1,args=('ADDRESS.1',))

sheet3 = pd.read_excel(r"Kelowna-Matt's Report II.xlsx",
                           sheetname=2,header=1, skiprows=(2,))
sheet3['Trip_Date'] = sheet3.apply(lambda x: date(2016,2,3), axis=1)
sheet3['Origin Address'] = sheet3.apply(Kelowna_Addresses,axis=1,args=('ADDRESS',))
sheet3['Dest Address'] = sheet3.apply(Kelowna_Addresses,axis=1,args=('ADDRESS.1',))

sheet4 = pd.read_excel(r"Kelowna-Matt's Report II.xlsx",
                           sheetname=3,header=1, skiprows=(2,))
sheet4['Trip_Date'] = sheet4.apply(lambda x: date(2016,2,4), axis=1)
sheet4['Origin Address'] = sheet4.apply(Kelowna_Addresses,axis=1,args=('ADDRESS',))
sheet4['Dest Address'] = sheet4.apply(Kelowna_Addresses,axis=1,args=('ADDRESS.1',))

sheet5 = pd.read_excel(r"Kelowna-Matt's Report II.xlsx",
                           sheetname=4,header=1, skiprows=(2,))
sheet5['Trip_Date'] = sheet5.apply(lambda x: date(2016,2,5), axis=1)
sheet5['Origin Address'] = sheet5.apply(Kelowna_Addresses,axis=1,args=('ADDRESS',))
sheet5['Dest Address'] = sheet5.apply(Kelowna_Addresses,axis=1,args=('ADDRESS.1',))

sheet6 = pd.read_excel(r"Kelowna-Matt's Report II.xlsx",
                           sheetname=5,header=1, skiprows=(2,))
sheet6['Trip_Date'] = sheet6.apply(lambda x: date(2016,2,6), axis=1)
sheet6['Origin Address'] = sheet6.apply(Kelowna_Addresses,axis=1,args=('ADDRESS',))
sheet6['Dest Address'] = sheet6.apply(Kelowna_Addresses,axis=1,args=('ADDRESS.1',))

sheet7 = pd.read_excel(r"Kelowna-Matt's Report II.xlsx",
                           sheetname=6,header=1, skiprows=(2,))
sheet7['Trip_Date'] = sheet7.apply(lambda x: date(2016,2,8), axis=1)
sheet7['Origin Address'] = sheet7.apply(Kelowna_Addresses,axis=1,args=('ADDRESS',))
sheet7['Dest Address'] = sheet7.apply(Kelowna_Addresses,axis=1,args=('ADDRESS.1',))

sheet8 = pd.read_excel(r"Kelowna-Matt's Report II.xlsx",
                           sheetname=7,header=1, skiprows=(2,))
sheet8['Trip_Date'] = sheet8.apply(lambda x: date(2016,2,9), axis=1)
sheet8['Origin Address'] = sheet8.apply(Kelowna_Addresses,axis=1,args=('ADDRESS',))
sheet8['Dest Address'] = sheet8.apply(Kelowna_Addresses,axis=1,args=('ADDRESS.1',))

sheet9 = pd.read_excel(r"Kelowna-Matt's Report II.xlsx",
                           sheetname=8,header=1, skiprows=(2,))
sheet9['Trip_Date'] = sheet9.apply(lambda x: date(2016,2,10), axis=1)
sheet9['Origin Address'] = sheet9.apply(Kelowna_Addresses,axis=1,args=('ADDRESS',))
sheet9['Dest Address'] = sheet9.apply(Kelowna_Addresses,axis=1,args=('ADDRESS.1',))

concat_df = pd.concat((sheet1,sheet2,sheet3,sheet4,sheet5,sheet6,sheet7,sheet8,sheet9))
concat_df.reset_index(drop=True).to_csv('Kelowna.csv',index_label='Trip_Number')
```
